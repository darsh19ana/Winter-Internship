```python
df['City_Mileage'] = df['City_Mileage'].str.replace('km/litre','')
df['City_Mileage'] = df['City_Mileage'].str.replace('?','')
df['City_Mileage'] = df['City_Mileage'].str.replace(',','.')
df.City_Mileage
```

pandas doesn't know how to convert such range from '12.5-12.7'(in this case) into a single float number. Thus, we created a function to convert the ranges into average to convert the column to float. 
This code defines a function convert_to_average that takes a string, splits it by the '-' character if present, converts the parts to floats, and returns their average. If the conversion fails (e.g., the string doesn't contain a numeric value), it returns NaN. Then it applies this function to the City_Mileage column. Finally, we are able to convert the column to float without issues.

```python
def convert_to_average(value):
    try:
        # Check if the value is a string
        if isinstance(value, str):
            # Split the string by '-' and convert the parts to float, then take the average
            parts = value.split('-')
            return sum(float(part) for part in parts) / len(parts)
        else:
            # If the value is not a string, just return it as is
            return value
    except:
        # If there is an error in conversion, return NaN
        return np.nan

# Apply the function to the 'City_Mileage' column
df['City_Mileage'] = df['City_Mileage'].apply(convert_to_average)

# Now you can safely convert to float
df['City_Mileage'] = df['City_Mileage'].astype(float)
```

## Data Visualization(Bivariate)

### 1. Correlation Matrix

```python
numeric_columns = df.select_dtypes(include=['number'])

correlation_matrix = numeric_columns.corr()
correlation_matrix
```

```python
plt.figure(figsize=(10, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='Spectral', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

From the above heatmap of the Correlation Matrix, we can see,
1. There is a strong positive correlation between the price of a car and its engine displacement. This suggests that cars with larger engines (more cc) tend to be more expensive.
2. Similar to displacement, there's a strong positive correlation between the number of cylinders in a car and its price. Cars with more cylinders are generally more expensive.
3. This strong positive correlation indicates that cars with more cylinders tend to have larger engine displacements.
4. Fuel tank capacity shows a moderate positive correlation with displacement (0.745281), cylinders (0.626478), and number of airbags (0.607564). This suggests that larger vehicles, which often have larger engines and more safety features, tend to have larger fuel tanks
5. The correlation of city mileage with other features is generally very low, indicating that city mileage does not vary strongly with these features.
6. Seating capacity has a negative correlation with price (-0.315518) and cylinders (-0.315835), suggesting that higher-priced cars or cars with more cylinders tend not to have higher seating capacity.
7. The correlation with basic warranty is generally low, although there's a modest correlation with displacement (0.238751) and fuel tank capacity (0.319723).
8. There is a  moderate positive correlations with displacement (0.550648), cylinders (0.444008), and fuel tank capacity (0.607564), suggesting that cars with larger engines and larger fuel tanks tend to have more airbags.
